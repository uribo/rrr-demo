```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')

penguins.head()
```

```python
sns.pairplot(penguins, hue='species', palette = 'muted')
plt.legend()
plt.title('Pairplot of penguins')
```
